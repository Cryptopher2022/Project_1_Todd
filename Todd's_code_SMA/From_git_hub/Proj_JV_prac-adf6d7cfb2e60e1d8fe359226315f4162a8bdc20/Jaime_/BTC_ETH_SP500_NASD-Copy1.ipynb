```python
# THIS IS THE BEGINNING OF THE CODE BLOCK FOR THE SIMPLE MOVING AVERAGE (SMA) FUNCTION

# Inputs required from User
# n = calendar days in the interval
# e.g., 2 months: n = 60, 6 months: n = 180

# window = number of days in the moving average
# e.g., 50 DMA: window = 50, 200 DMA: window = 200



n = 180 # 6 month data block from today
window = 25 # 25 day moving average (DMA)

btc_data = pd.read_csv("./Resources/btc-usd.csv",
                       index_col='Date', 
                        parse_dates=True, 
                        infer_datetime_format=True
                      )
btc_data
```

```python
from datetime import date
today = str(date.today())
print(today)
```

```python
# from the initial user input n is already set

from datetime import datetime, timedelta
import time
today = datetime.today()
n = 180
start_date = today - timedelta(days=n)
#print(f"Today's date,"  str(today))
#)
#print(f"The start date for time period,"  start_date.strftime('%Y-%m-%d')
#)
print (today)
print (start_date)
```

```python
btc_df =btc_data.loc[start_date:today]

btc_df.head(85)
```

```python
# View header and tail of BTC for 2 months
print("BTC-USD data frame from: April 20, 20220 - May 05, 2022")
print("-----------------------------------------------------")
display(btc_df.head())
display(btc_df.tail())
```

```python
# Dropping last 2 columns of BTC 
btc_new_df = btc_df.drop(columns=['Open','High','Low','Volume'], axis=1)
print("BTC-USD update:")
print("---------------")
btc_new_df
```

```python
"""
This is the code block for the 15 day moving average for Bitcoin over a 2 month period.  

"""

end = 60
n <= end
n = 0
window = 15
n_window = n + window
btc_avg = btc_new_df

while n in range(16,59):
    #valid==(n < 60)
    n_window = n + window
    time = btc_new_df[n:n_window].sum()
    avg = time/window
    btc_avg = btc_avg.append(avg, ignore_index=False)
    n += 1 

    
btc_final = btc_avg.drop_duplicates()
print("This is the listing of 15 day moving averages for Bitcoin")
display(btc_final)
```

```python
btc_final.hvplot(x='Date', y='Close', kind = 'line', value_label = "This is the moving average of Bitcoin over the user specific timeframe and window")
```
