```python
def BTC_SMA():
#    end = 60
#    n <= end
#    n = 0
#    window = 15
    n_window = n + window
    btc_avg = btc_new_df

    while n in range(16,59):
        #valid==(n < 60)
        n_window = n + window
        time = btc_new_df[n:n_window].sum()
        avg = time/window
        btc_avg = btc_avg.append(avg, ignore_index=False)
        n += 1 

    
    btc_final = btc_avg.drop_duplicates()
    print("This is the listing of 15 day moving averages for Bitcoin")
    display(btc_final)
    
    return btc_final
```
